```python
# cargar y unir dos datasets
data_ejecutivos=pd.read_excel("/content/sample_data/Taller2_ETL.xlsx",sheet_name="Ejecutivos")
data_ejecuciones=pd.read_excel("/content/sample_data/Taller2_ETL.xlsx",sheet_name="BaseEjecuciones")
```

```python
# merge de datos
ejecutivos_ejecuciones=pd.merge(data_ejecutivos,data_ejecuciones,how="inner",on="CEDULA")
ejecutivos_ejecuciones
```
